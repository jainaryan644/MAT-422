3.2.1 Limits and Continuity

Limits describe the behavior of a function as its input approaches a certain value.
Key concepts:
- Limit of a function f(x) as x approaches a:  We write lim (x->a) f(x) = L,
meaning that as x gets arbitrarily close to a, the values of f(x) get arbitrarily close to L.

- Continuity: A function is continuous at a point if the limit of the function at that point exists and is equal to the function's value at that point.

- One-sided limits: These are limits as x approaches a from either the left or the right.
- Infinite limits: These are limits that approach positive or negative infinity.
- Asymptotes: These are lines that a function approaches but never touches.
- Continuity and differentiability: A differentiable function is always continuous, but the converse is not true.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example 1: Limit of a function
def f(x):
  return x**2

x = np.linspace(-5, 5, 100)
y = f(x)

plt.plot(x, y)
plt.title("Limit of f(x) = x^2 as x approaches 2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.scatter(2, f(2), color='red', label="f(2) = 4")
plt.legend()
plt.show()

# As x approaches 2, f(x) approaches 4. This demonstrates the limit of the function.

# Example 2: Continuity
def g(x):
  if x == 0:
    return 1
  else:
    return x

x = np.linspace(-5, 5, 100)
y = [g(val) for val in x]

plt.plot(x, y)
plt.title("Continuity at x = 0")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.scatter(0, g(0), color='red', label="g(0) = 1")
plt.legend()
plt.show()

# The function g(x) is not continuous at x = 0 because the limit as x approaches 0 is 0, but g(0) = 1.

# Example 3: One-sided limits
def h(x):
  if x >= 0:
    return x
  else:
    return -x

x = np.linspace(-5, 5, 100)
y = [h(val) for val in x]

plt.plot(x, y)
plt.title("One-sided limits at x = 0")
plt.xlabel("x")
plt.ylabel("h(x)")
plt.show()

# The limit as x approaches 0 from the left is 0, and the limit as x approaches 0 from the right is 0.
# However, the function is continuous at x = 0.


# Example 4: Infinite Limits
def k(x):
  return 1/x

x = np.linspace(-5, 5, 100)
y = [k(val) if val != 0 else np.nan for val in x]

plt.plot(x, y)
plt.title("Infinite Limit at x = 0")
plt.xlabel("x")
plt.ylabel("k(x)")
plt.show()

# As x approaches 0, k(x) approaches positive or negative infinity depending on the direction.
```

3.2.2 Derivatives

Derivatives measure the instantaneous rate of change of a function. They represent the slope of the tangent line to the function at a specific point. Key concepts include:

- Instantaneous rate of change: The derivative calculates how a function's output changes as its input changes infinitesimally.
- Slope of the tangent line: The derivative's value at a point equals the slope of the line that touches the function's curve at that point.
- Differentiability: A function is differentiable at a point if its derivative exists at that point. This means the function is smooth and has a defined tangent line at that point.

```python
def f(x):
  return x**2

def derivative(f, x, h=0.001):
  return (f(x + h) - f(x)) / h

x = np.linspace(-5, 5, 100)
y = f(x)

# Calculate the derivative at a specific point
x_point = 2
derivative_at_x = derivative(f, x_point)

# Plot the function and its tangent line
plt.plot(x, y, label="f(x) = x^2")

tangent_line_y = f(x_point) + derivative_at_x * (x - x_point)
plt.plot(x, tangent_line_y, label="Tangent line at x = {}".format(x_point))

plt.scatter(x_point, f(x_point), color='red', label="Point (x, f(x))")

plt.title("Derivative and Tangent Line")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()

print("The derivative of f(x) at x = {} is approximately {}".format(x_point, derivative_at_x))

# The slope of the tangent line at x = 2 represents the instantaneous rate of change of f(x) at that point.
```

3.2.3 Taylor's Theorem


Taylor's Theorem is a fundamental result in calculus that provides a way to approximate the value of a function using its derivatives at a specific point.

**Key Concepts:**

1. **Polynomial Approximation:** Taylor's theorem states that a sufficiently differentiable function can be approximated by a polynomial, known as the Taylor polynomial, centered at a particular point.

2. **Remainder Term:** The approximation is not exact, and the difference between the actual function value and the Taylor polynomial value is called the remainder term.

3. **Convergence:** The accuracy of the approximation depends on the degree of the Taylor polynomial and the distance from the center point. As the degree of the polynomial increases, the approximation typically becomes more accurate, and the remainder term decreases.

4. **Applications:** Taylor's theorem has widespread applications in various fields, including physics, engineering, and computer science. It is used to solve differential equations, analyze the behavior of functions, and perform numerical computations.


**In essence, Taylor's Theorem allows us to represent a complex function with a simpler polynomial near a specific point, providing a powerful tool for understanding and approximating function behavior.**


```python


def taylor_series_exp(x, n_terms=5):
  result = 0
  for i in range(n_terms):
    result += (x**i) / np.math.factorial(i)
  return result

# Example usage e^x:
x_values = np.linspace(-3, 3, 100)
actual_y = np.exp(x_values)
approx_y = [taylor_series_exp(x, n_terms=10) for x in x_values]

plt.plot(x_values, actual_y, label="e^x")
plt.plot(x_values, approx_y, label="Taylor Series Approximation (10 terms)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Taylor Series Approximation of e^x")
plt.legend()
plt.show()
```
